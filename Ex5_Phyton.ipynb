```python
'''Crie uma função que receba um nome, separe as 4 primeiras letras e complete com a
palavra mon.
Ex: Daniel → Danimon
'''

def nome_modificado(nome):
  return nome[:4] + 'mon'

print(nome_modificado('Thauan'))
print(nome_modificado(input('Digite seu nome: ')))
```

```python
'''Crie uma função que receba um texto e retorne o texto com as seguintes alterações:
Retorne o texto com cada palavra invertida.
Exemplo: Usuário digitou “Aula de Python”
Sistema retornou “aluA ed nohtyP”
'''
def inversor(palavra):
  return palavra[::-1]

def inversor_frase(frase):
  palavras = frase.split()

  frase_invertida = ''
  for palavra in palavras:
    frase_invertida += inversor(palavra) + ' '
  return frase_invertida[:-1]

print(inversor_frase(input('Digite sua frase: ')))
```

```python
'''Crie uma função que receba um texto e retorne o texto com as seguintes alterações:
Faça o deslocamento de 3 letras para direita por palavra
Exemplo: Usuário digitou “Aula de Python”
Sistema retornou “ulaA ed honPyt”
'''

def criptografar(palavra, deslocamento):
  desloc_real = deslocamento % len(palavra)
  return palavra[-desloc_real:] + palavra[:-desloc_real]

def criptografar_frase(frase, deslocamento=3):
  palavras = frase.split()

  frase_invertida = ''
  for palavra in palavras:
     frase_invertida += criptografar(palavra, deslocamento) + ' '
  return frase_invertida[:-1]

print(criptografar_frase('Aula de Python',deslocamento=4))
frase = input('Digite a frase: ')
deslocamento_usuario = int(input('Digite quantas letras deslocar: '))
print(criptografar_frase(frase, deslocamento_usuario))
```
